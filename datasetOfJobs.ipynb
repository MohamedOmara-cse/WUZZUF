```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

>Reading data set and showing first 5 Rows:

```python
df = pd.read_csv('Wuzzuf_Jobs.csv')
df.head(5)
```

```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.info();
```

```python
Jobs_companies = df['Company'].value_counts()[:5]
print(Jobs_companies)
```

```python
plt.subplots(figsize=(30,10))
companies = ['Confidential','Mishkat Nour', 'Expand Cart', 'EGIC', 'Aqarmap.com']
plt.pie(Jobs_companies, labels=companies )

plt.legend();
```

```python
x = df['Title'].value_counts()[:10]
print(x)
```

```python
plt.subplots(figsize=(25,7))
plt.bar(x.index,x,width = 0.4, color='blue');
plt.title("MOST POPULAR JOBS")
plt.xlabel("Jobs")
plt.ylabel("Demand");
```

```python
y = df['Location'].value_counts()[:10]
print(y)
```

```python
plt.subplots(figsize=(15,7))
plt.bar(y.index,y,width = 0.4, color='blue');
plt.title("MOST POPULAR LOCATIONS")
plt.ylabel("RATE")
plt.xlabel("CITIES");
```

```python
z = df['Skills'].value_counts()[:5]
print(z)
```

```python
plt.subplots(figsize=(30,10))
groups = ['group1','group2', 'group3', 'group4', 'group5']
plt.pie(z, labels=groups )
plt.legend();
```
